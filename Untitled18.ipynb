```python
# Import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import mean_squared_error, accuracy_score, f1_score
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

# Missing Values

```python
# Check for missing values
print(data.isnull().sum())

# Fill missing values or drop rows/columns as necessary
# Fill missing values in numerical columns with their mean
numerical_cols = data.select_dtypes(include=np.number).columns
data[numerical_cols] = data[numerical_cols].fillna(data[numerical_cols].mean())

# For categorical columns, filling in with the mode strategy was used
# For categorical columns:
categorical_cols = data.select_dtypes(include='object').columns
data[categorical_cols] = data[categorical_cols].fillna(data[categorical_cols].mode().iloc[0])

# Verify that there are no more missing values
print("\nMissing values after handling:")
print(data.isnull().sum())
```

## Visualizing Distribution of Water Quality Indicators

```python
# Show key numerical water quality indicators for visualization
water_quality_indicators = ['pH', 'TH', 'TDS', 'WQI']

# Create histograms for the selected indicators
plt.figure(figsize=(15, 10))
for i, col in enumerate(water_quality_indicators):
    plt.subplot(2, 2, i + 1)
    sns.histplot(data=data, x=col, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# Encode Categorical Variables

```python
# Encode categorical variables using Label Encoding
label_encoder = LabelEncoder()
data['Water Quality Classification'] = label_encoder.fit_transform(data['Water Quality Classification'])
```

# Split Features and Target Variables

```python
# Define features and target variables
X = data.drop(['WQI', 'Water Quality Classification', 'Well_ID'], axis=1)
y_regression = data['WQI']
y_classification = data['Water Quality Classification']
```

# Split the Data into Training and Test Sets

```python
# Split the dataset into training and test sets
X_train, X_test, y_train_reg, y_test_reg, y_train_class, y_test_class = train_test_split(
    X, y_regression, y_classification, test_size=0.2, random_state=42
)
```

# Feature Scaling

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline

# Identify numerical and categorical columns (excluding 'Well_ID' which was dropped earlier)
numerical_cols = X.select_dtypes(include=np.number).columns
categorical_cols = X.select_dtypes(include='object').columns

# Create a ColumnTransformer to apply different transformations to different columns
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ],
    remainder='passthrough' # Keep other columns (if any) - though none expected after selecting num and cat
)

# Apply the preprocessing
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

# Now X_train_processed and X_test_processed are numpy arrays with all numerical data, ready for modeling
# X_train_processed = pd.DataFrame(X_train_processed, columns=preprocessor.get_feature_names_out())
# X_test_processed = pd.DataFrame(X_test_processed, columns=preprocessor.get_feature_names_out())

print("Processed X_train shape:", X_train_processed.shape)
print("Processed X_test shape:", X_test_processed.shape)
```

## **Build the Neural Network Model**

# The Regression Model for WQI

```python
# Build the regression model
regression_model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train_processed.shape[1],)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)  # Output layer for regression
])

regression_model.compile(optimizer='adam', loss='mean_squared_error')
```

# Classification Model for Water Quality Classification

```python
# Build the classification model
classification_model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train_processed.shape[1],)),
    layers.Dense(32, activation='relu'),
    layers.Dense(5, activation='softmax')  # Output layer for classification (5 classes in your dataset)
])

classification_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

# **Train the Models**

```python
# Train the regression model
regression_model.fit(X_train_processed, y_train_reg, epochs=100, batch_size=32, validation_split=0.2)

# Train the classification model
classification_model.fit(X_train_processed, y_train_class, epochs=100, batch_size=32, validation_split=0.2)
```

# **Evaluate the Models**

# Evaluate the Regression Model

```python
# Predict and evaluate the regression model
y_pred_reg = regression_model.predict(X_test_processed)
mse = mean_squared_error(y_test_reg, y_pred_reg)
print(f'Mean Squared Error for WQI: {mse}')
```

# Evaluate the Classification Model

```python
# Predict and evaluate the classification model
y_pred_class_probs = classification_model.predict(X_test_processed)
y_pred_class = np.argmax(y_pred_class_probs, axis=1)

accuracy = accuracy_score(y_test_class, y_pred_class)
f1 = f1_score(y_test_class, y_pred_class, average='weighted') # Use weighted average for multi-class

print(f'Accuracy for Water Quality Classification: {accuracy}')
print(f'F1-score for Water Quality Classification: {f1}')
```

```python
from sklearn.metrics import classification_report

# Evaluate the model on the test set
print("\nEvaluating the model...")
loss, accuracy = classification_model.evaluate(X_test_processed, y_test_class, verbose=0)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")

# Generate predictions and display a classification report
y_pred_probs = classification_model.predict(X_test_processed)
y_pred_classes = np.argmax(y_pred_probs, axis=1)

# Inverse transform predicted and true labels to get original class names
y_test_labels = label_encoder.inverse_transform(y_test_class)
y_pred_labels = label_encoder.inverse_transform(y_pred_classes)

report = classification_report(y_test_labels, y_pred_labels, target_names=label_encoder.classes_)
print("\nClassification Report:")
print(report)

# Make a prediction on a sample of the test data
print("\nMaking a prediction on a sample of the test data...")
# Get the first 5 samples from the processed test data
sample_data_processed = X_test_processed[:5]
predictions = classification_model.predict(sample_data_processed)
predicted_classes = np.argmax(predictions, axis=1)
predicted_labels = label_encoder.inverse_transform(predicted_classes)

# Show the predicted labels for the first 5 samples.
print(f"\nPredicted Water Quality Classification for the sample:")
for i, label in enumerate(predicted_labels):
    print(f"Sample {i+1}: {label}")
```
